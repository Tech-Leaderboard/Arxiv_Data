```python
import urllib
import feedparser
import numpy as np
import pandas as pd
```

```python
url = 'http://export.arxiv.org/api/query?search_query=all:AI&start=0&max_results=10000'
```

```python
feedparser._FeedParserMixin.namespaces['http://a9.com/-/spec/opensearch/1.1/'] = 'opensearch'
feedparser._FeedParserMixin.namespaces['http://arxiv.org/schemas/atom'] = 'arxiv'
```

```python
# perform a GET request using the base_url and query
response = urllib.request.urlopen(url).read()
# parse the response using feedparser
feed = feedparser.parse(response)
```

```python
data = []
authors = set()
```

```python
for entry in feed.entries:
    author_string = entry.authors
    sub_date = entry.updated
    pub_date = entry.published
    title = ""
    for ch in entry.title:
        if ch != '\n':
            title += ch
    for author in author_string:
        row = []
        if author in authors:
            print(author)
        authors.add(author)
        row.append(author.name)
        row.append(sub_date)
        row.append(pub_date)
        row.append(title)
        data.append(row)
```

```python
with open('sample.csv', 'w') as outfn:
    s = 'author_name' + ',' + 'submission_date' + ',' + 'publication_date' + ',' + 'title'
    outfn.write(s)
    outfn.write('\n')
    for idx in range(len(data)):
        s = ",".join(data[idx])
        outfn.write(s)
        outfn.write('\n')
```
